```python
import pandas as pd
```

```python
file_path = "Resources/School_enrolment_and_demographics_2021_22.xlsx"
```

```python
Schools_df = pd.read_excel(file_path)
```

```python
Schools_df.head(3)
```

```python
#Schools and Demographics Dataframe

Columns = Schools_df.columns
print(Columns)
```

```python
#School_Location Table

School_location_Columns = ['School Number', 'Street', 'Municipality', 'City', 'Province', 'Postal Code', 'Fax Number', 'Latitude', 'Longitude']

School_location_df = Schools_df[School_location_Columns]

School_location_df.head(3)
```

```python
# Cleaning School_location_df 

School_location_df.loc[:, :] = School_location_df.fillna(" ")
```

```python
#Creating a CSV for the school location 

School_location_df.to_csv('Output Tables/school_location_data.csv', index=False)
```

```python
# Creating the Schools Table

School_Info_Columns = ['School Number' , 'Board Number' , 'School Name' , 'School Type', 'School Level' , 'Phone Number' , 'School Website']

School_Info_df = Schools_df[School_Info_Columns]

School_Info_df.head()
```

```python
# Cleaning School_Info_df 

cleaned_School_Info_df = School_Info_df.fillna(" ")
```

```python
#Creating a CSV for the school location 

cleaned_School_Info_df.to_csv('Output Tables/school_info_data.csv', index=False)
```

```python
# Creating the Board Table

Board_Info_Columns = ['Board Number', 'Board Name', 'Board Type','Municipality','Board Website']

Board_Info_df = Schools_df[Board_Info_Columns]

Board_Info_df.head()
```

```python
# Cleaning Board_Info_df 

Cleaned_Board_Info_df = Board_Info_df.fillna(" ")
```

```python
#Creating a CSV for School Boards

Cleaned_Board_Info_df.to_csv('Output Tables/Board_info_data.csv', index=False)
```

```python
# Enrolment tables Preparation 

Enrolment_file_path = "Resources/enrol_grade_21_22_en.txt"

Enrolment_df = pd.read_csv(Enrolment_file_path, delimiter="|")

Enrolment_df.head(3)
```

```python
#Schools and Demographics Dataframe

Enrolment_Columns = Enrolment_df.columns

print(Enrolment_Columns)
```

```python
# Creating the Board Table

Enrolment_by_grade_Columns = ['School Number', 
       'School Type', 'School Level',
       'Grade 1 Enrolment', 'Grade 2 Enrolment', 'Grade 3 Enrolment',
       'Grade 4 Enrolment', 'Grade 5 Enrolment', 'Grade 6 Enrolment',
       'Grade 7 Enrolment', 'Grade 8 Enrolment', 'Grade 9 Enrolment',
       'Grade 10 Enrolment', 'Grade 11 Enrolment', 'Grade 12 Enrolment']

Enrolment_by_grade_df = Enrolment_df[Enrolment_by_grade_Columns]

Enrolment_by_grade_df.head()
```

```python
#Elementary school enrolment table

elementary_schools_enrolment_df = Enrolment_by_grade_df.loc[Enrolment_by_grade_df["School Level"] == "Elementary" , ["School Number"	, 
                                                                                                                    "School Type" , 
                                                                                                                    "Grade 1 Enrolment",
                                                                                                                    "Grade 2 Enrolment"	,
                                                                                                                    "Grade 3 Enrolment"	,
                                                                                                                    "Grade 4 Enrolment"	,
                                                                                                                    "Grade 5 Enrolment"	,
                                                                                                                    "Grade 6 Enrolment"	,
                                                                                                                    "Grade 7 Enrolment"	,
                                                                                                                    "Grade 8 Enrolment"]]

elementary_schools_enrolment_df.head()
```

```python
#Secondary school enrolment table

Secondary_schools_enrolment_df = Enrolment_by_grade_df.loc[Enrolment_by_grade_df["School Level"] == "Secondary" , ['School Number' , 'School Type' , 'Grade 9 Enrolment','Grade 10 Enrolment', 'Grade 11 Enrolment', 'Grade 12 Enrolment']]

Secondary_schools_enrolment_df.head()
```

```python
#Delating the NaN fields and "<"
#Extract the table for elementary_schools_enrolment_df

elementary_schools_enrolment_df.fillna(0, inplace=True)

elementary_schools_enrolment_df.replace("<",'', regex=True, inplace=True)

elementary_schools_enrolment_df.to_csv('Output Tables/Elementary_School_Enrolment.csv', index=False)
```

```python
#Delating the NaN fields and "<"
#Extract the table for Secondary_schools_enrolment_df

Secondary_schools_enrolment_df.fillna(0, inplace=True)

Secondary_schools_enrolment_df.replace("<",'', regex=True, inplace=True)

Secondary_schools_enrolment_df.to_csv('Output Tables/Secondary_School_Enrolment.csv', index=False)
```
