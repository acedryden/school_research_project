```python
import pandas as pd
```

```python
file_path = "Resources/new_sif_data_table_2021_22prelim_en_january2024.xlsx"
```

```python
Schools_df = pd.read_excel(file_path)
```

```python
Schools_df.head(3)
```

```python
#Schools and Demographics Dataframe

Columns = Schools_df.columns
print(Columns)
```

```python
#School_Location Table

School_location_Columns = ['School Number', 'Board Number' , 'School Name' , 'School Type', 'School Level' , 'Street', 'Municipality', 'City', 'Province', 'Postal Code', 'Fax Number', 'Latitude', 'Longitude', 'Phone Number' , 'School Website']

Schools_all_df = Schools_df[School_location_Columns]

Schools_all_df.head(3)
```

```python
# List of school numbers to be deleted
school_numbers_to_delete = ['SC9205', 'SC9206', 'SC8201', 'SC0302', 'SC0501', 'SC0701', 'SC0403', 'SC9801', 'SC9901', 'SC6103', 'SC0505', 'SC0506', 'SC0802', 'SC0901']

# Delete records with specified school numbers
Schools_all_df = Schools_all_df[~Schools_all_df['School Number'].isin(school_numbers_to_delete)]

# Display the first 3 rows of the updated DataFrame
Schools_all_df.head(3)
```

```python
#Creating a CSV for the school location 

Schools_all_df.to_csv('Output Tables/school_data.csv', index=False)
```

```python
# Creating the Board Table

Board_Info_Columns = ['Board Number', 'Board Name', 'Board Type','Municipality','Board Website']

Board_Info_df = Schools_df[Board_Info_Columns]

Board_Info_unique_df = Board_Info_df.drop_duplicates(subset=['Board Number'])

Board_Info_unique_df
```

```python
#Creating a CSV for School Boards

Board_Info_unique_df.to_csv('Output Tables/Board_data.csv', index=False)
```

```python
# Enrolment tables Preparation 

Enrolment_file_path = "Resources/enrol_grade_21_22_en.txt"

Enrolment_df = pd.read_csv(Enrolment_file_path, delimiter="|")

Enrolment_df.head()
```

```python
#Schools and Demographics Dataframe

Enrolment_Columns = Enrolment_df.columns

print(Enrolment_Columns)
```

```python
# Creating the Board Table

Enrolment_by_grade_Columns = ['School Number', 
       'School Type', 'School Level',
       'Grade 1 Enrolment', 'Grade 2 Enrolment', 'Grade 3 Enrolment',
       'Grade 4 Enrolment', 'Grade 5 Enrolment', 'Grade 6 Enrolment',
       'Grade 7 Enrolment', 'Grade 8 Enrolment', 'Grade 9 Enrolment',
       'Grade 10 Enrolment', 'Grade 11 Enrolment', 'Grade 12 Enrolment']

Enrolment_by_grade_df = Enrolment_df[Enrolment_by_grade_Columns]

Enrolment_by_grade_df.head()
```

```python
# Eliminating the NaN values

Enrolment_by_grade_df = Enrolment_by_grade_df.fillna(0)

Enrolment_by_grade_df.replace("<",'', regex=True, inplace=True)

Enrolment_by_grade_df.head()
```

```python
# Adding total enrolment for the year

Enrolment_by_grade_df[['Grade 1 Enrolment', 'Grade 2 Enrolment', 'Grade 3 Enrolment', 'Grade 4 Enrolment', 
                      'Grade 5 Enrolment', 'Grade 6 Enrolment', 'Grade 7 Enrolment', 'Grade 8 Enrolment', 
                      'Grade 9 Enrolment', 'Grade 10 Enrolment', 'Grade 11 Enrolment', 'Grade 12 Enrolment']] = Enrolment_by_grade_df[['Grade 1 Enrolment', 'Grade 2 Enrolment', 'Grade 3 Enrolment', 'Grade 4 Enrolment', 
                                                                                                              'Grade 5 Enrolment', 'Grade 6 Enrolment', 'Grade 7 Enrolment', 'Grade 8 Enrolment', 
                                                                                                              'Grade 9 Enrolment', 'Grade 10 Enrolment', 'Grade 11 Enrolment', 'Grade 12 Enrolment']].apply(pd.to_numeric, errors='coerce')


Enrolment_by_grade_df['Total Enrolment'] = Enrolment_by_grade_df[['Grade 1 Enrolment', 'Grade 2 Enrolment', 'Grade 3 Enrolment', 'Grade 4 Enrolment', 
                                                                 'Grade 5 Enrolment', 'Grade 6 Enrolment', 'Grade 7 Enrolment', 'Grade 8 Enrolment', 
                                                                 'Grade 9 Enrolment', 'Grade 10 Enrolment', 'Grade 11 Enrolment', 'Grade 12 Enrolment']].sum(axis=1)
```

```python
# Adding the year

Enrolment_by_grade_df.insert(1, "Year", "2021-2022")

Enrolment_by_grade_df.head()
```

```python
#extracting the table 

Enrolment_by_grade_df.to_csv('Output Tables/Enrolment_Data.csv', index=False)
```

```python
#Graduation Table

Graduation_file_path = "Resources/graduation_rate_by_Board 2021-2022.xlsx"

Graduation_df = pd.read_excel(Graduation_file_path)

Graduation_df.head()
```

```python
School_path = "Output Tables/school_data.csv"

School_df = pd.read_csv(School_path)

School_df.head()
```

```python
Enrolment_path = "Output Tables/Enrolment_Data.csv"

Enrolment_df = pd.read_csv(Enrolment_path)

Enrolment_df.head()
```

```python
# Create a list of unique school numbers from both tables
missing_from_df1 = Enrolment_df[~Enrolment_df["School Number"].isin(School_df["School Number"])]
missing_from_df2 = School_df[~School_df["School Number"].isin(Enrolment_df["School Number"])]

print("School numbers missing from Enrolment_df:")
print(missing_from_df1)

print("\nSchool numbers missing from School_df:")
print(missing_from_df2)
```

```python
missing_from_Enrolment_df = Enrolment_df[~Enrolment_df["School Number"].isin(School_df["School Number"])]["School Number"]
missing_from_School_df = School_df[~School_df["School Number"].isin(Enrolment_df["School Number"])]["School Number"]

print("School numbers missing from Enrolment_df:")
print(missing_from_Enrolment_df)

print("\nSchool numbers missing from School_df:")
print(missing_from_School_df)
```

```python
School_df = School_df[~School_df["School Number"].isin(missing_from_School_df)]
Enrolment_df = Enrolment_df[~Enrolment_df["School Number"].isin(missing_from_Enrolment_df)]

print("Updated School_df:")
print(School_df)

print("\nUpdated Enrolment_df:")
print(Enrolment_df)
```

```python
School_df.to_csv('Output Tables/school_data.csv', index=False)
```

```python
Enrolment_df.to_csv('Output Tables/Enrolment_Data.csv', index=False)
```
