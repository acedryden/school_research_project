## Enrolment for the next five year

```python
# Reading csv file for enrolment
df = pd.read_csv("enrolment_data_final.csv")
```

```python
# Extracting the first year for each row
df["Year"] = df["Year"].str.split("-").str[0]

# Transform the Year column to int
df["Year"] = df["Year"].astype(int)
```

```python
# Identify unique schools
unique_schools = df['School Number'].unique()

# Define the range of future years
future_years = range(2022, 2027)

# Create an empty list to store future enrolment data
future_enrolment_data = []

# Iterate over each school to generate predictions
for school in unique_schools:
    # Filter data for the current school
    school_data = df[df['School Number'] == school]

    # Iterate over each grade's enrolment data
    for grade in range(1, 13):

        # Calculate the average growth rate
        grade_column = f'Grade {grade} Enrolment'
        enrolment_growth_rate = school_data[grade_column].pct_change().mean()

        # Assuming the last known year's enrolment is the base for forecasting
        last_known_year = school_data['Year'].max()
        last_year_data = school_data[school_data['Year'] == last_known_year]

        # Get the enrolment for the last known year for the current grade
        last_year_enrolment = last_year_data[grade_column].values[0]

        # Iterate over future years
        for year in future_years:

            # Project future enrolment for the current grade using the average growth rate
            predicted_enrolment = last_year_enrolment * (1 + enrolment_growth_rate)

            # Append the predicted enrolment for the current school, grade, and year
            future_enrolment_data.append({
                'School Number': school,
                'Year': year,
                'Grade': grade,
                'Enrolment': predicted_enrolment
            })

            # Update the last year enrolment for the next iteration
            last_year_enrolment = predicted_enrolment

# Create a DataFrame from the list of future enrolment data
future_df = pd.DataFrame(future_enrolment_data)
```

```python
# Pivot the DataFrame to have separate columns for each grade's enrolment
future_df_pivot = future_df.pivot_table(index=['School Number', 'Year'], columns='Grade', values='Enrolment').reset_index()

# Renaming the columns
future_df_pivot.columns = ['School Number', 'Year'] + [f'Grade {grade} Enrolment' if grade else 'Year' for grade in future_df_pivot.columns[2:]]
future_df_pivot.columns.name = None
```

```python
# Renaming the future_df_pivot
future_df = future_df_pivot

# Replace infinite values (inf) with NaN
future_df.replace([np.inf, -np.inf], np.nan, inplace=True)

# Drop rows containing NaN values
future_df.fillna(0, inplace=True)

# Dimentioning the Grade columns as integers
columns = ['Grade 1 Enrolment', 'Grade 2 Enrolment', 'Grade 3 Enrolment', 'Grade 4 Enrolment',
           'Grade 5 Enrolment', 'Grade 6 Enrolment', 'Grade 7 Enrolment', 'Grade 8 Enrolment',
           'Grade 9 Enrolment', 'Grade 10 Enrolment', 'Grade 11 Enrolment', 'Grade 12 Enrolment']
for col in columns:
    future_df[col] = pd.to_numeric(future_df[col], errors='coerce').astype(int)
```

```python
future_df.sample(10)
```

```python
# Extracting the future enrolment data
future_df.to_csv("Future_Enrolment.csv", index = False)
```
