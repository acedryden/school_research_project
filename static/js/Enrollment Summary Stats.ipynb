```python
import psycopg2
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
from scipy.stats import linregress
```

```python
enrollment_url = "Enrolment_Data - Enrolment_Data.csv"
enrollment_data = pd.read_csv(enrollment_url) 
enrollment_data.head()
```

```python
print(enrollment_data.columns)
```

```python
columns_to_select = ['School Type', 'School Level', 'Total Enrolment']
summary_enrollment = enrollment_data[columns_to_select].copy()

summary_enrollment
```

```python
summary_stats = summary_enrollment.groupby(['School Type', 'School Level']).agg({'Total Enrolment': 'sum'}).reset_index()
summary_stats
```

```python
summary_stats_json = summary_stats.to_json(orient='records')
print(summary_stats_json)
```
